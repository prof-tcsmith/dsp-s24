```python
# Importing basic libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

## Section 1: Demonstrating Data Imbalance

### 1.1 Creating an Imbalanced Dataset

```python
# Create an imbalanced dataset
X, y = make_classification(
    n_classes=2, class_sep=1.5,
    weights=[0.1, 0.9],
    n_informative=2, n_redundant=0,
    flip_y=0, n_features=2,
    n_clusters_per_class=1, n_samples=1000, random_state=10
)
```

```python
# Plotting the dataset
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label="Minority class", alpha=0.6)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label="Majority class", alpha=0.6)
plt.title("Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### Observations
* Discussion about how the imbalance is visible
* The potential impact on a predictive model


## Section 2: Resampling Techniques


### 2.1 Random Under-Sampling

```python
# Code to perform random under-sampling
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

```python
# Plotting the dataset
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Minority class", alpha=0.6)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Majority class", alpha=0.6)
plt.title("Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### 2.2 Random Over-Sampling

```python
# Code to perform random over-sampling
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
```

```python
# Plotting the dataset
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Minority class", alpha=0.6)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Majority class", alpha=0.6)
plt.title("Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### 2.3 Synthetic Minority Over-Sampling Technique (SMOTE)

```python
# Code to perform SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Plotting the dataset
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Minority class", alpha=0.6)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Majority class", alpha=0.6)
plt.title("Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### 2.4 Adaptive Synthetic Sampling (ADASYN)

```python
# Code to perform ADASYN
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(random_state=42)
X_res, y_res = adasyn.fit_resample(X, y)
```

```python
# Plotting the dataset
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Minority class", alpha=0.6)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Majority class", alpha=0.6)
plt.title("Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

## Section 3: Resampling Techniques Conclusion

#### 3.1 Random Under-Sampling

* **Conditions for Preference:**
  * When you have a large dataset and reducing its size won't impact the model's performance significantly.
  * When computational resources are limited.
* **Caveats**:
  * Risk of losing important data points from the majority class.
  * Might lead to underfitting.

#### 3.2 Random Over-Sampling

* **Conditions for Preference**:
  * When the minority class is exceptionally small.
  * When adding more instances would make the model more robust.
* **Caveats**:
  * Risk of overfitting due to duplicate instances.
  * Increase in computational cost.

### 3.3 SMOTE (Synthetic Minority Over-sampling Technique)

* **Conditions for Preference**:
  * When random over-sampling leads to overfitting.
  * When under-sampling leads to loss of important data.
  * When the dataset is high-dimensional.
* **Caveats**:
  * Synthetic instances may not always be meaningful.
  * Computationally more expensive than random over-sampling.


### 3.4 ADASYN (Adaptive Synthetic Sampling)

* **Conditions for Preference**:
  * When you want to adaptively generate minority data instances according to their distribution.
  * When SMOTE isn't capturing the complexity of the data.
* **Caveats**:
  * Similar to SMOTE, but generally more computationally intensive due to the adaptive nature.
